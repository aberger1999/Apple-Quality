```python
import pandas as pd

# Load the file into a DataFrame
apple_quality = pd.read_csv("apple_quality.csv")

apple_quality.head()
```

```python
# Get the dimensions of the dataset
dimensions = apple_quality.shape
print("Number of rows:", dimensions[0])
print("Number of columns:", dimensions[1])
```

```python
print(apple_quality.isnull().sum())
```

```python
apple_quality_cleaned = apple_quality.dropna()

print(apple_quality_cleaned.isnull().sum())
```

```python
apple_quality_cleaned
```

```python
apple_quality_cleaned.columns
```

```python
# Label encoding: 1 = good, 0 = bad
binary_labels = [1 if quality == "good" else 0 for quality in apple_quality_cleaned['Quality']]
apple_quality_cleaned['Quality'] = binary_labels
apple_quality_cleaned
```

```python
#The feature A_id adds no value to the model and can be dropped
apple_quality_cleaned.drop("A_id", axis=1, inplace=True)
```

```python
numerical_columns = apple_quality_cleaned.select_dtypes(include=['float64', 'int64'])

for column in numerical_columns.columns:
    sns.histplot(apple_quality_cleaned[column], bins=10, kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
sns.countplot(data=apple_quality_cleaned, x='Quality')
plt.title('Countplot for Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(apple_quality_cleaned.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
